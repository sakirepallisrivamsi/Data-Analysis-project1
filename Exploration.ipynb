```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cleaned_dataset.csv')
```

```python
# correlation heatmap 
# method for correlation can be either 'kendall' , 'spearman', 'pearson'
f, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(df.corr(method='kendall'), cmap="YlGnBu", annot=True, ax=ax)
plt.show()
```

### PCA

```python
# Principal component Analysis
# n_components is the output dimensions
# The variance indicates what percentage of information that can be maximized by choosing n columns of data.
from sklearn.decomposition import PCA
columns = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8', 'X9', 'X10', 'X11', 'X12', 'X13', 'score', 'Y']
X = df[columns]
for i in range(1,16):
    pca = PCA(n_components=i)
    pca.fit(X)
    print("variance with {} components:".format(i),sum(pca.explained_variance_ratio_))
```
