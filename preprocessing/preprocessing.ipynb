```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset_uncleaned_28072020.csv')
```

```python
df= df[[ 'industry', 'disbursal_date',
       'tnc_amount_first', 'tnc_ir_first', 'tnc_tenure_first',
       'min_cibil_score', 'loan_type',  'd_date_of_birth', 'vintage',
       '3m_avg_bal', '3m_avg_bal_new', 'total_liab_new', '90adb',
       'l_created_date', 'tenure', 'score', 'rounded_score']]
```

```python
# only the selected columns are used because the other extra columns were used to query data effectively (data retrieval filters)
```

```python
df.head()
```

```python
# inserting null values with the mean of the column
df = df.fillna(df.mean())
```

```python
# converting date columns to type datetime
df['disbursal_date']=pd.to_datetime(df.disbursal_date)
df['d_date_of_birth'] = pd.to_datetime(df.d_date_of_birth)
df['l_created_date'] = pd.to_datetime(df.l_created_date)

```

```python
# choosing numerical columns only for analysis (excluding categorical) 
columns = ['disbursal_date', 'tnc_amount_first',
       'tnc_ir_first', 'tnc_tenure_first', 'min_cibil_score',
        'd_date_of_birth', 'vintage', '3m_avg_bal', '3m_avg_bal_new',
       'total_liab_new', '90adb', 'l_created_date', 'tenure']
```

```python
# normalizing data via min-max
for j in columns:
    normalized_disbursal_list = []
    max_disbursal = df[j].max() 
    min_disbursal = df[j].min()
    for i in range(df.shape[0]):
        normalized_disbursal = (df[j][i] - min_disbursal)/(max_disbursal- min_disbursal)
        normalized_disbursal_list.append(normalized_disbursal)
    df[j] = normalized_disbursal_list
```
